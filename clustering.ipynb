# Applying a clustering algorithm to the data

Here we use a Gaussian Mixture Model Algorithm

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c1 = pd.read_csv('./Data/c1_fe.csv', index_col='cycle')
c2 = pd.read_csv('./Data/c2_fe.csv', index_col='cycle')
c3 = pd.read_csv('./Data/c3_fe.csv', index_col='cycle')
c4 = pd.read_csv('./Data/c4_fe.csv', index_col='cycle')
```

```python
# Here I select the columns for the GMM, excluding the cycles one
gmm_data = c1.drop(columns=['SOH'])

# I create a Gaussian Mixture Model with 20 clusters
num_components = 20
gmm = GaussianMixture(n_components=num_components, random_state=0)

# Fit the GMM to my data
gmm.fit(gmm_data)

# Predict the cluster labels for each sample
cluster_labels = gmm.predict(gmm_data)

# Add the cluster labels to the DataFrame
c1['Cluster'] = cluster_labels

# Plot the clusters
plt.figure(figsize=(20, 10))
sns.scatterplot(x='cycle', y='Q', hue='Cluster', data=c1)
plt.title("GMM Clusters")
plt.xlabel("cycle")
plt.ylabel("Q")
plt.show()
```

This is one way to apply a clustering algorithm. This one wasn't fully designed, but it is here just as an example.
